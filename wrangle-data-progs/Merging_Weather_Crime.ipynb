# Import Weather Data

```python
weather_zip = zipfile.ZipFile("C:/Users/Barbieri/DSC-Program/DC-Criminalistics/weather-data/crime-weather-data.zip",
                              mode='r')

weather_json = weather_zip.open('crime-weather-data.json')

weather_dict = json.load(weather_json)
```

```python
weather_list = []

for weather in weather_dict:
    try:
        latitude = weather['latitude']
    except KeyError:
        latitude = ''

    try:
        longitude = weather['longitude']
    except KeyError:
        longitude = ''

    try:
        time = weather['currently']['time']
    except KeyError:
        time = ''

    try:
        summary = weather['currently']['summary']
    except KeyError:
        summary = ''

    try:
        icon = weather['currently']['icon']
    except KeyError:
        icon = ''
        
    try:
        precip_intensity = weather['currently']['precipIntensity']
    except KeyError:
        precip_intensity = ''
        
    try:
        precip_probability = weather['currently']['precipProbability']
    except KeyError:
        precip_probability = ''
        
    try:
        temp = weather['currently']['temperature']
    except KeyError:
        temp = ''

    try:
        app_temp = weather['currently']['apparentTemperature']
    except KeyError:
        app_temp = ''

    try:
        dew_point = weather['currently']['dewPoint']
    except KeyError:
        dew_point = ''
        
    try:
        humidity = weather['currently']['humidity']
    except KeyError:
        humidity = ''
        
    try:
        pressure = weather['currently']['pressure']
    except KeyError:
        pressure = ''
        
    try:
        wind_speed = weather['currently']['windSpeed']
    except KeyError:
        wind_speed = ''
        
    try:
        wind_gust = weather['currently']['windGust']
    except KeyError:
        wind_gust = ''
        
    try:
        wind_bearing = weather['currently']['windBearing']
    except KeyError:
        wind_bearing = ''
        
    try:
        cloud_cover = weather['currently']['cloudCover']
    except KeyError:
        cloud_cover = ''        

    try:
        uv_index = weather['currently']['uvIndex']
    except KeyError:
        uv_index = ''

    try:
        visibility = weather['currently']['visibility']
    except KeyError:
        visibility = ''
    
    weather_list.append([latitude,longitude,time,
                         temp,app_temp,precip_probability,
                         humidity,cloud_cover,uv_index,
                         summary,icon,precip_intensity,
                         dew_point,pressure,wind_speed,
                         wind_gust,wind_bearing,cloud_cover,visibility])
```

```python
columns = ['latitude','longitude','time',
           'temp','app_temp','precip_probability',
           'humidity','cloud_cover','uv_index',
           'summary','icon','precip_intensity',
           'dew_point','pressure','wind_speed',
           'wind_gust','wind_bearing','cloud_cover','visibility']

weather_df = pd.DataFrame(data=weather_list, columns=columns)
```

# Merge Crime and Weather Data

### Compare Dataframes

Notice that there are 13,967 more instances of crime than there are of the weather data. it will be important to note which instances of crime are missing.

```python
weather_df_len = len(weather_df)
crime_df_len = len(crime_df)

print(weather_df_len, crime_df_len, weather_df_len-crime_df_len)
```

### Clean Weather Data

```python
#There are missing rows.
weather_df = weather_df[weather_df['time'] != '']

#There are negative time values.
weather_df = weather_df[weather_df['time'] > 0]
```

```python
#Convert time to match crime data.
def converttime(row):
    time = datetime.datetime.fromtimestamp(row).strftime('%Y-%m-%dT%H:%M:%S.000')

    return time
```

```python
weather_df['c_time'] = weather_df['time'].apply(converttime)
```

```python
weather_df.head()
```

```python
crime_weather_merge = crime_df.merge(weather_df,
                                     how='left',
                                     left_on=['LATITUDE','LONGITUDE','START_DATE'],
                                     right_on=['latitude','longitude','c_time'])
```
