```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create dataframe for each year's csv of crime data from DC Open Data
# Source link for 2018: http://opendata.dc.gov/datasets/crime-incidents-in-2018
df_dc_2009 = pd.read_csv('Crime_Incidents_in_2009.csv')
df_dc_2010 = pd.read_csv('Crime_Incidents_in_2010.csv')
df_dc_2011 = pd.read_csv('Crime_Incidents_in_2011.csv')
df_dc_2012 = pd.read_csv('Crime_Incidents_in_2012.csv')
df_dc_2013 = pd.read_csv('Crime_Incidents_in_2013.csv')
df_dc_2014 = pd.read_csv('Crime_Incidents_in_2014.csv')
df_dc_2015 = pd.read_csv('Crime_Incidents_in_2015.csv')
df_dc_2016 = pd.read_csv('Crime_Incidents_in_2016.csv')
df_dc_2017 = pd.read_csv('Crime_Incidents_in_2017.csv')
df_dc_2018 = pd.read_csv('Crime_Incidents_in_2018.csv')
```

```python
# Check headings and data
# Can quickly see that BID contains nulls and is therefore likely not useful for us
# We'll keep it for now though
df_dc_2009.head()
```

```python
# Check headings and data. Data from 2009 and 2018 have the same headers.
# This should mean we can put all csvs together into one dataframe
df_dc_2018.head()
```

```python
df_dc_full = pd.concat([df_dc_2009,df_dc_2010,df_dc_2011,df_dc_2012,df_dc_2013,df_dc_2014,df_dc_2015,df_dc_2016,df_dc_2017,df_dc_2018])
```
