###### This script combines the ACS 5 year Census data pulled by block group
Note the directory paths needed for reading and writing the files.  Note, ACS 5 year detailed tables are not available at the block group level for the years 2010, 2011, and 2012.

```python
import pandas as pd
```

```python
data09 = pd.read_csv('Clean/BlockGroupClean/2009.txt')
data09['Year'] = 2009

data13 = pd.read_csv('Clean/BlockGroupClean/2013.txt')
data13['Year'] = 2013

data14 = pd.read_csv('Clean/BlockGroupClean/2014.txt')
data14['Year'] = 2014

data15 = pd.read_csv('Clean/BlockGroupClean/2015.txt')
data15['Year'] = 2015

data16 = pd.read_csv('Clean/BlockGroupClean/2016.txt')
data16['Year'] = 2016

data17 = pd.read_csv('Clean/BlockGroupClean/2017.txt')
data17['Year'] = 2017
```

```python
frames = [data09, data13, data14, data15, data16, data17]
```

```python
df_bg = pd.concat(frames)
```

```python
df_bg.columns = ['TotalPop', 'TPopMargin', 'UnWgtSampleCtPop', 'PerCapitaIncome', 'PerCapIncMargin', 'MedianHouseholdInc', 'MedHouseholdIncMargin', 'MedianAge', 'MedianAgeMargin', 'HousingUnits', 'HousingUnitsMargin', 'UnweightedSampleHousingUnits', 'State', 'County', 'Tract', 'BlockGroup', 'Year']
```
