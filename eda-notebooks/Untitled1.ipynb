```python
import os
import pandas as pd
import sqlite3
```

```python
def checkDirectory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
    else:
        print('Directory already exists, exiting.')
```

```python
def importCensusData(inputFilePath, columns):
    """
    Import Census data and return a dataframe.
    """
    data = []

    with open(inputFilePath, 'r') as infile:
        lines = infile.readlines()

        for index, line in enumerate(lines):
            if index == 0:
                columns = line.replace('],', '').replace('[', '').replace('\"', '').replace(']]', '').replace('\n','').split(',')
            else:
                row = line.replace('],', '').replace('[', '').replace('\"', '').replace(']]', '').replace('\n','').split(',')
                data.append(row)

    census_df = pd.DataFrame(data=data, columns=columns)

    return census_df
```
